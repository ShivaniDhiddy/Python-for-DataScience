```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import numpy as np
import re
import pandas as pd
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
names=[]
color=[]
prices=[]
ratings=[]
no_ratings=[]
ram=[]
rom=[]
screen_sz=[]
cam1=[]
cam2=[]
cam2_f=[]
battery=[]
Processor=[]
brand=[]
```

```python
for page in range(1,14):
    url="https://www.flipkart.com/mobiles/mi~brand/pr?sid=tyy%2C4io&otracker=nmenu_sub_Electronics_0_Mi&sort=price_asc&page="+str(page)
    driver.get(url)
    content = driver.page_source
    soup = BeautifulSoup(content, "html.parser")
    
    for a in soup.findAll('a', href=True, attrs={'class': '_31qSD5'}):
        # for name of mobile
        product=a.find("div",{"class":'_3wU53n'})
        name=product.text.split("(")[0]
        brand.append("Mi")
        
        
        # price
        price=a.find("div", {"class":"_1vC4OE _2rQ-NK"})
        p_1=str(price.text).replace(",","")
        price=p_1.strip("₹")
        
        #rating
        rating=a.find("div", {"class":"hGSR34"})
        if rating:
            ratings.append(rating.text)
        else:
            ratings.append(np.nan)
        # number of ratings
        norate=a.find("span", {"class":"_38sUEc"})
        if norate:
            norate=str(re.search(r'^\d+.?\d+.?\d+\s',norate.text).group()).strip().replace(",","")
            no_ratings.append(norate)
        else:
            no_ratings.append(np.nan)
    
        # colour
        colour=str(re.search(r'(\w+\s?\w+?),',product.text).group()).strip(",")
    
        # memory storage
    
        mem=(a.find("div",{"class":'_3ULzGw'})).find("li",{"class":"tVe95H"})
        mem_ram=str(re.search("^\d\s\w+",mem.text).group()).replace(" GB","")
        mem_rom=str(re.search("(\d*)\s+GB",mem.text[11:]).group()).replace(" GB","")
    
        #screen size
        ss=a.find("ul",{"class":'vFw0gD'})
        details=ss.text.strip()
        ss=str(re.search("(\d+.\d*)\s+cm",details).group()).replace(" cm","")
    
        # camera
        cam_p=str((re.search(r"\w+\d+MP",details).group()).replace("Display","")).replace("MP","")
        
        #front_cam
        if details:
            cam_f=re.findall(r'(\d+MP)\s\+\s\d+MP\sDual|(\d+MP)\sFront',details)
            #print(cam_f)
            if cam_f:
                cam2.append(cam_f[0][1])
                if cam_f[0][0]:
                    cam2.append(cam_f[0][0])
            else:
                cam2.append(np.nan)
        
        #battery
        bat=(re.search(r"\d{4}",details)).group()
        battery.append(bat)
        
        #processor
        if details:
            x1=details.split("Battery")[1]
            if "Processor" in details:
                Processor.append(x1.split("Processor")[0])
            else:
                Processor.append(np.nan)

    
        
        
        names.append(name)
        prices.append(price)
        color.append(colour)
        #ratings.append(rating.text)
        #no_ratings.append(norate)
        ram.append(mem_ram)
        rom.append(mem_rom)
        screen_sz.append(ss)
        cam1.append(cam_p)
        
for i in cam2:
    if i:
        cam2_f.append(i)   
            
        
```
